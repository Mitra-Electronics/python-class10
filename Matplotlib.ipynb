```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,6,7,8,9]
plt.plot(x, y)
plt.show()
```

```python
x1 = [1,2,3,4,5]
y1 = [2,6,7,8,9]
x2 = [1,2,3,4,5]
y2 = [2,6,7,8,9]
plt.plot(x1, y1)
plt.plot(x2, y2, marker='o', linestyle='-', color='r')
plt.show()
```

```python
sp = [100,200,300,400,500]
op = [200,600,700,800,900]
plt.plot(x1, y1, marker='o', linestyle='-', color='r')
plt.xlabel("Selling Price")
plt.ylabel("Original Price")
plt.title("SP vs OP")
plt.show()
```

```python
sp = [100,200,300,400,500]
op = [200,600,700,800,900]
plt.plot(x1, y1, marker='o', linestyle='-', color='r')
plt.xlabel("Selling Price")
plt.ylabel("Original Price")
plt.title("SP vs OP")
plt.grid() # Adds an overlay of grids
plt.show()
```

```python
x1 = [1,2,3,4,5]
y1 = [2,6,7,8,9]
x2 = [1,2,3,4,5]
y2 = [2,6,7,8,9]
plt.plot(x1, y1, label="First line")
plt.plot(x2, y2, marker='o', linestyle='-', color='r', label="Second line")
plt.legend() #plt.plot(label="*") Labels necessary for legend to be shown
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,6,7,8,9]
plt.scatter(x, y, marker="x", color="red", label="Scatter")
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,6,7,8,9]
plt.bar(x, y, color="red", label="Scatter")
plt.legend()
plt.show()
```
